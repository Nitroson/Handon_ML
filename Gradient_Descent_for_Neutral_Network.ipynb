```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df[['age' , 'bought_insurance']] , df.bought_insurance , test_size = 0.2 , random_state = 25)
```

## Scaling

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age']/100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age']/100
```

```python
model = keras.Sequential([
    keras.layers.Dense(1 , input_shape = (2,) , activation = 'sigmoid' , kernel_initializer = 'ones' , bias_initializer = 'zeros')
])

model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])

model.fit(X_train_scaled , y_train , epochs = 500)
```

```python
model.evaluate(X_test_scaled , y_test)
```

## Implementing using python

```python
def sigmoid(x):
  import math
  return 1/(1+math.exp(-x))
```

```python
def prediction_function(age , bought_insurance):
  weighted_sum = coef[0]*age + coef[1]* bought_insurance + intercept
  return sigmoid(weighted_sum)
```

```python
def log_loss(y_true , y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i , epsilon) for i in y_predicted]
  y_predicted_new = [min(i , 1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(x):
  return 1/(1+np.exp(-x))

sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age , bought_insurance , y_true , epochs , loss_thresold):
  w1 = w2 = 1
  bias = 0
  rate = 0.5
  n = len(age)

  for i in range(epochs):
    weighted_sum = w1 * age + w2 * bought_insurance + bias
    y_predicted = sigmoid_numpy(weighted_sum)

    loss = log_loss(y_true , y_predicted)
    (y_predicted - y_true)

    w1d = (1/n)*np.dot(np.transpose(age) , (y_predicted - y_true))
    w2d = (1/n)*np.dot(np.transpose(bought_insurance) , (y_predicted - y_true))

    bias_d = np.mean(y_predicted - y_true)
    w1 = w1 - rate * w1d
    w2 = w2 - rate * w2d
    bias = bias - rate * bias_d

    print(f'Epochs : {i} , w1 : {w1} , w2 : {w2} , bias : {bias} , loss : {loss}')
    if loss <= loss_thresold:
      break


  return w1 , w2 , bias
```
